```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Baca gambar berwarna
image_color = cv2.imread('TEST_RAMA.jpg')
```

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

```python
# Terapkan median filtering pada gambar berwarna
median_filtered_color = cv2.medianBlur(image_color, 5)
```

```python
# Terapkan mean filtering (blur) pada gambar grayscale
mean_filtered_gray = cv2.blur(image_gray, (5, 5))
```

```python
# Tampilkan gambar asli berwarna dan hasil filtering
plt.figure(figsize=(15, 5))
```

```python
# Tampilkan gambar asli berwarna dan hasil filtering
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(median_filtered_color, cv2.COLOR_BGR2RGB))
plt.title('After Filter Median')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(mean_filtered_gray, cmap='gray')
plt.title('Mean Filtered (Grayscale)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image (for Mean Filtered)')
plt.axis('off')

plt.show()
```
